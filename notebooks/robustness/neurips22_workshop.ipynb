```python
K=1.5
for local_budget in [1, 2, 3, 4, "deg+0", "deg+2"]:
    count_all = 0
    count_overrobust = 0
    count_adversarial_only = 0
    for label, K, exp in experiment_manager.experiment_iterator("l2-weak", ["GCN"], [K]):  
        f_wrt_y = exp.robustness_statistics["c_gnn_robust_when_both"]
        g_wrt_y = exp.robustness_statistics["c_bayes_robust_when_both"]
        for deg in f_wrt_y.keys():
            for f_wrt_y_rob, g_wrt_y_rob in zip(f_wrt_y[deg], g_wrt_y[deg]):
                if local_budget == "deg+0":
                    budget = int(deg)
                elif local_budget == "deg+2":
                    budget = int(deg)+2
                else:
                    budget = local_budget
                if f_wrt_y_rob > budget:
                    f_wrt_y_rob = budget
                if g_wrt_y_rob > budget:
                    g_wrt_y_rob = budget
                if f_wrt_y_rob > g_wrt_y_rob:
                    count_overrobust += 1
                else:
                    count_adversarial_only += 1
                count_all += 1
    print(f"Local Budget: {local_budget}")
    print(f"Cases of Over-Robustness: {count_overrobust} of {count_all} i.e. {count_overrobust/count_all*100:.2f}%")
    print(f"Cases Adv only: {count_adversarial_only} of {count_all} i.e. {count_adversarial_only/count_all*100:.2f}%")
```
